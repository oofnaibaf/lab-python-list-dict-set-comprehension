````python
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory
````

````python
# ALL INVENTORY FUNCTIONS START
def init_inventory(products):
 #1 Initializes inventory dictionary based on user input. Combined to a single line
    return {product: int(input(f"Enter the quantity for {product}: ")) for product in products}

def update_inventory(customer_orders, inventory):
#6 UPdates the updated inventory. Ensuring quantities don't go below 0
    for order in customer_orders:
        if order in inventory:
            inventory[order] = max(inventory[order] - 1, 0)  # Reduce by 1, with a floor at 0
        else:
            print(f"Warning: {order} is not available in inventory.")

def print_inventory(inventory, title="Updated Inventory"):
#3 Prints inventory
    print(title)
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
# ALL INVENTORY FUNCTIONS END

# ALL CUSTOMER FUNCTIONS START
def get_customer_orders():
    #2 Prompts the user to enter product names and returns a set of customer orders.
    return {input("Enter the product name to order (or type 'done' to finish): ") 
            for _ in iter(lambda: input(""), 'done')}

def calc_order_stats(customer_orders, products):
   #Calculates order statistics: total products ordered and percentage of unique products ordered.
    total_ordered = len(customer_orders)
    unique_products_ordered = len(customer_orders.intersection(products))
    percentage_unique = (unique_products_ordered / total_ordered * 100) if total_ordered else 0
    return total_ordered, percentage_unique

def print_order_stats(order_stats):
    #Prints the order statistics.
    total_ordered, percentage_unique = order_stats
    print(f"Total products ordered: {total_ordered}")
    print(f"Percentage of unique products ordered: {percentage_unique:.2f}%")
# ALL CUSTOMER FUNCTIONS END

# Main execution sequence
def main():
    products = ["Apple", "Banana", "Orange", "Grapes"]
    inventory = init_inventory(products)
    customer_orders = get_customer_orders()
    update_inventory(customer_orders, inventory)
    print_order_stats(calc_order_stats(customer_orders, products))
    print_inventory(inventory)

# Run the program
if __name__ == "__main__":
    main()
````
